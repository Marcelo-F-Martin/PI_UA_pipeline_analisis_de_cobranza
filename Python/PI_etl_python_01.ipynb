```python
# SE IMPORTAN LIBRERIAS DE TRABAJO
#-----------------------------------

import pandas as pd
import requests
import io
```

```python
# CONEXION AL REPOSITORIO FUENTE (SIMULANDO DIRECTORIO LOCAL)
#-----------------------------------

url_raw = "https://github.com/Marcelo-F-Martin/PI_UA_pipeline_analisis_de_cobranza/releases/download/v1.0.0/Detalle.de.comisiones.Broker.xls"
response = requests.get(url_raw)
df_raw = pd.read_excel(io.BytesIO(response.content),engine='xlrd',header=None)
print(df_raw.iloc[:15, :10])
##print(response.status_code)
##response.raise_for_status()
```

```python
# BUSQUEDA DE PALABRA CLAVE PARA CREACION DE DF
#-----------------------------------

encabezado_clave = "Periodo"
indice_encabezado_fila = df_raw[df_raw.apply(lambda row: encabezado_clave in row.astype(str).values, axis=1)].index

fila_inicio = indice_encabezado_fila[0]
print(f"✅ Encabezado clave '{encabezado_clave}' encontrado en la Fila (índice 0): {fila_inicio}")
```

```python
# 2a. Aislar la fila del encabezado encontrado
#--------------------------------------

fila_encabezado = df_raw.iloc[fila_inicio]

# 2b. 'col_inicio' será el índice de la columna donde se encuentra la clave
# Esto es crucial para ignorar las columnas añadidas a la izquierda.

col_inicio = fila_encabezado[fila_encabezado.astype(str) == encabezado_clave].index[0]
print(f"  ✅ Columna de inicio de datos (índice 0): {col_inicio}")
```

```python
# 2c. Re-leer el DataFrame usando la fila encontrada como encabezado
# (Se necesita la ruta completa del archivo, que es una variable externa al bucle interno)
# Se necesita un DataFrame limpio, no el bruto, para el recorte final.

# Por eficiencia, si ya tienes el df_bruto, lo mejor es usarlo para el recorte:
df_raw.columns = df_raw.iloc[fila_inicio] # Asignar la fila encontrada como nuevo encabezado
df_clean = df_raw[fila_inicio + 1:].reset_index(drop=True) # Eliminar filas superiores

#hay un paso intermedio preparado para borrar columnas a la izquierda del df_raw. Por ahora se omite eso.

df_final_hoja = df_clean.copy()
df_final_hoja
```
